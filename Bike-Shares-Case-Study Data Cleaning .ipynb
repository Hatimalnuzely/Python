```python
# Importing the data

trips4_apr21 = pd.read_csv('/Users/hatimalnuzely/Desktop/Case Study Bike-Share/202104-divvy-tripdata.csv')
trips5_may21 = pd.read_csv('/Users/hatimalnuzely/Desktop/Case Study Bike-Share/202105-divvy-tripdata.csv')
trips6_jun21 = pd.read_csv('/Users/hatimalnuzely/Desktop/Case Study Bike-Share/202106-divvy-tripdata.csv')
trips7_jul21 = pd.read_csv('/Users/hatimalnuzely/Desktop/Case Study Bike-Share/202107-divvy-tripdata.csv')
trips8_aug21 = pd.read_csv('/Users/hatimalnuzely/Desktop/Case Study Bike-Share/202108-divvy-tripdata.csv')
trips9_sep21 = pd.read_csv('/Users/hatimalnuzely/Desktop/Case Study Bike-Share/202109-divvy-tripdata.csv')
trips10_oct21 = pd.read_csv('/Users/hatimalnuzely/Desktop/Case Study Bike-Share/202110-divvy-tripdata.csv')
trips11_nov21 = pd.read_csv('/Users/hatimalnuzely/Desktop/Case Study Bike-Share/202111-divvy-tripdata.csv')
trips12_dec21 = pd.read_csv('/Users/hatimalnuzely/Desktop/Case Study Bike-Share/202112-divvy-tripdata.csv')
trips1_jan22 = pd.read_csv('/Users/hatimalnuzely/Desktop/Case Study Bike-Share/202201-divvy-tripdata.csv')
trips2_feb22 = pd.read_csv('/Users/hatimalnuzely/Desktop/Case Study Bike-Share/202202-divvy-tripdata.csv')
trips3_mar22 = pd.read_csv('/Users/hatimalnuzely/Desktop/Case Study Bike-Share/202203-divvy-tripdata.csv')
```

```python
# Creating list for all 12 dataframes

trips_list = [
    trips4_apr21, trips5_may21, trips6_jun21, trips7_jul21, trips8_aug21, trips9_sep21, trips10_oct21,
    trips11_nov21, trips12_dec21, trips1_jan22, trips2_feb22, trips3_mar22
]
```

```python
trips_list
```

```python
# Concatenating all dataframes into one dataframe

data = pd.concat(trips_list, axis= 0 , ignore_index=True)
data.columns
```

```python
# Taking a copy in case Something went wrong

data_copy = data.copy()
data_copy.head()
```

```python
data_copy.describe()
```

```python
# Converting columns data type from object to datetime

data_copy['started_at'] = pd.to_datetime(data_copy['started_at'])
data_copy['ended_at'] = pd.to_datetime(data_copy['ended_at'])


data_copy.info()
```

```python
# Checking for columns with nulls

data_copy.isnull().sum()
```

```python
# Checking for rows with nulls

data_copy[data_copy.isnull().any(axis=1)]
```

```python
# Removing null values

data_copy = data_copy.dropna()
```

```python
data_copy.isnull().any(axis=1).sum()
```

```python
# Checking for Duplicate 

data_copy.duplicated().sum()
```

```python
# Adding new column called 'Ride_Length' for calculating the length of each ride by subtracting 
# the column 'started_at' from the column 'ended_at'

data_copy['ride_length'] = data_copy['ended_at'] - data_copy['started_at']
```

```python
# Adding new column called 'ride_Length_minutes' for calculating the length in minutes

ride_Length_minutes = data_copy['ended_at'] - data_copy['started_at']
```

```python
# Convert 'ride_length' to minutes

data_copy['ride_Length_minutes'] = np.round(ride_Length_minutes.dt.total_seconds()/60, 2)
```

```python
data_copy.info()
```

```python
data_copy.head()
```

```python
# Creating a new column called 'day_of_week' to calculate the day of the week that each ride started

data_copy['day_of_week'] = data_copy['started_at'].dt.day_name()
```

```python
data_copy.head()
```

```python
data_copy.info()
```

```python
# Adding new column 'month' to calculate the month

data_copy['month'] = data_copy['started_at'].dt.month_name()
```

```python
data_copy.head()
```

```python
# Checking for negative values in 'ride_Length_minutes' column

(data_copy['ride_Length_minutes']<0).sum()
```

```python
# Assigning dataframe to a variable to get rid of the negative values

data_clean = data_copy
```

```python
# Assign the positive values to the new variable

data_clean = data_copy[data_copy['ride_Length_minutes']>0]
```

```python
# Convert variable to data frame

data_clean = pd.DataFrame(data_clean)
```

```python
# Checking new dataframe for negative values

(data_clean['ride_Length_minutes']< 0).sum()
```

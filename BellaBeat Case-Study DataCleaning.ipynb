```python
# Making a copy of the data if something went wrong

activity_data = activity.copy()
sleep_data = sleep.copy()
```

```python
# Converting ActivityDate and SleepDay data types into date data types

activity_data['ActivityDate'] = pd.to_datetime(activity_data['ActivityDate'], format="%m/%d/%Y")
sleep_data['SleepDay'] = pd.to_datetime(sleep_data['SleepDay'])
```

```python
# Renaming date columns to merge 

activity_data.rename(columns={'ActivityDate':'date_time'}, inplace=True)
```

```python
sleep_data.rename(columns={'SleepDay':'date_time'}, inplace=True)
```

```python
# Merging data frames 

data = pd.merge(activity_data, sleep_data, on=['Id','date_time'])
```

```python
# Removing Duplicates 

data.drop_duplicates(inplace=True)
```

```python
# Taking a copy of data

data_copy = data.copy()
```

```python
# Adding a new column 'DayOfWeek'

data_copy['DayOfWeak'] = data_copy['date_time'].dt.day_name()
```

```python
data_copy
```

```python
# Adding a new column 'Total_mins'

data_copy['Total_mins'] = data_copy['VeryActiveMinutes']+data_copy['FairlyActiveMinutes']+
data_copy['SedentaryMinutes']+ data_copy['LightlyActiveMinutes']
```

```python
# Adding a new column 'Total_hours'


data_copy['Total_hours'] = round(data_copy['Total_mins'] / 60)
```
